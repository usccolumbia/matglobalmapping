```python
import numpy as np
import pandas as pd
import glob
import os


allmp = pd.read_csv("../../input_data/targets.csv")  
allmp=allmp.reset_index(drop=True)

directory = "../../input_data/poscar/feature_topo_compo/"
i =0
topo_feature = []
for file in glob.iglob(f'{directory}/*'):
    #print(file)
    filename =file.split('/')[-1]
    #print(filename)
    mpid = filename.split('_')[0]
    #print(mpid)
    Ftopo = np.load(file)
    #print(len(Ftopo))

    entry = [mpid] + list(Ftopo)
    
    
    topo_feature.append(entry)
    
    if i%10000 ==0:
        df = pd.DataFrame(topo_feature)
        print(df.shape)
        df.to_csv('tmp_topo_test.csv',index=False)
        #break
    i = i+1
    #os.system(f"python cif2vasp.py {filename}")
df = pd.DataFrame(topo_feature)
#print(df.head)
df.to_csv('topo_feature.csv',index=False)    
```
