```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import manifold
import pandas as pd
import os
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import matplotlib as mpl
```

```python
def plot_BG(S_data):
    X=S_data.loc[:]['x']
    Y=S_data.loc[:]['y']
    plt.scatter(X,Y,marker='*',color='blue',s=15,zorder=2)
```

```python
def plot_MP_BG(S_lowDWeights):
    plt.cla()#清除当前图形中的当前活动轴,所以可以重复利用
    S_data=pd.DataFrame({'x':S_lowDWeights[:,0],'y':S_lowDWeights[:,1]})
    S_data.columns = ['x', 'y']
    #print(S_data.head)
    X=S_data.loc[:]['x']
    Y=S_data.loc[:]['y']
    plt.scatter(X,Y,marker='o',color='white',s=20,zorder=1, edgecolor="grey")
```
